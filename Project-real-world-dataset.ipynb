### The phenomenon

Psychological resilience is a valuable, arguably essential trait for any third level student to possess. Whilst third level education is intended to be a very exciting and enjoyable time in a persons life, students typically endure some very gruelling and stressful academic challenges, especially from an emotional and psychological point of view. The manner in which a student copes with the pressures of working toward attaining a first class undergraduate or postgraduate qualification will have notable effects on their mental wellbeing and final year grade. Every student employs their own coping styles, whether it's meditation, watching Netflix, listening to music, talking to friends or even alcohol consumption. The way in which each student copes, whether they adopt adaptive or maladaptive strategies will often have reciprocal effects on their academic performance, and ultimately their emotional and psychological wellbeing. These factors should not be ignored. Psychological resilience can act as a crucial buffer, protecting best outcomes for students. But what exactly is psychological resilience and how does it relate to student performance?

#### Pychological resilience

Psychological resilience can be succinctly defined as the ability to successfully cope with a crisis and to return to pre-crisis status quickly [1]. Resilience is apparent when the person uses mental processes and behaviours in promoting personal assets and protecting oneself from the potential negative effects of stressors. Essesntially, psychological resilience exists in people who develop psyhological and behavioural capabilities that allow them to reamin calm and objective during crises and to move forward from these events without long-term negative consequences. Resilience is a positive adaption after a stressful or adverse situation. When a person is overwhelmed by daily stress, it typically disrupts their internal and external sense of balance which can present both challenges and opportunities. It is not just about overcoming deeply stressful situations but also about emerging from them with competent functioning. Resilience is best understood as a process more than a trait. It is a cycle, a reactional process to stressful situations followed by an adaptive or maladaptive behaviour. Indeed, stressors can cause students to erupt with anger, implode with negative emotions or go numb, or simply become upset about the disruptive change. The first two reactions tend to lead people to play the victim by blaming others and rejecting any coping methods. These individuals instinctively *react* rather than *respond*. The third reaction type promotes wellbeing and is employed by resilient people who become upset about the disruptive state, yet change their current position to cope with the issue. These individuals tend to step back, make realistic plans, have confidenece in their own strengths, have good problem solving skills and have a higher degree of emotional regulation than those who choose to submit to natural reactions of fear, anger and helplessness [2]. The phenomenon exists among all individuals, student or not, yet warrants particular attention among third level students.

This study aims to explore some of the most important variables, including resilience, which affect and relate to student performance in an Irish Institute of Technology. The students have recently completed their Higher Diplomas and represent two different course cohorts: Those who studied Data Analytics and those who studied Software Development.

### Exploration of variables

Let's first take a look at the Age variable in the dataset. The Age variable is a continuous numerical variable. For both student cohorts, their ages range from 17 to 65. Age distributions can vary depending on the type of populations in the study. For example most population pyramids will display a trapezoidal or pyramid shaped distibution. After some consideration, for students ranging from 17 to 65 years, it was deemed appropriate to incorporate a blend of uniform and triangular distribution in order to reperesent the samples correctly. Both  distribution parameters allow you to set both the lower and upper ranges of age, but the triangular distribution also features a mode which is useful, i.e. the value which appears most often (the peak) to be set as well [3].

```python
# Import the necessary modules:

import pandas as pd
import numpy as np

# Set the seed so the variable values are reproducible:

np.random.seed(1234) 


# Setting the age range and sample size of students by concatenating the uniform and triangular distributions.
# This method will generate a convincing student age distribution [3]. 

Age1 = np.random.uniform(17,65,125)  # (low, high, size)
Age2 = np.random.triangular(17,37,65,125) # (left, mode, right, size)
Age = np.concatenate((Age1,Age2)) # Concatenating the distributions

Age.round(0)
```

#### Relationships among student variables

Exploring some of the relationships between the variables should offer greater insight into the dynamics at play among IT students. 

First we will look at the covariance values of the variables. Covariance and correlation describe how two variables are related in a dataset. Essentially, the student variables are inferred as positively related if they move in the same direction, or negatively related if they move in opposite directions. Covariance indicates how two variables are related, i.e. positively or negatively, and it measures variables that have different *units* of measurement. However, it is not possible to measure the *degree* to which the variables move together because covariance does not use one standard unit of measurement. Therefore, to measure the degree to which variables move together, we use correlation. Correlation standardizes the measure of *interdependence* between any two variables and, consequently, tells us how closely the two variables move. The correlation measurement (coefficient), always takes on a value between 1 and – 1 [9]. 

In order to calculate and then evaulate the correlations among the student variables, we first need to synthesize covariance values that offer a close approximate simulation of the true relationships of the variables. To do this we can stay with the numpy.random package, and employ the numpy.random.multivariate_normal distribution function. 

The multivariate normal distribution is often used to describe, at least approximately, any set of (possibly) correlated real-valued *random variables* each of which clusters around a mean value [10]. Therefore, this distributional function is an ideal candidate for simulating covariance values and then determining the correlations between the student variables.
The random.multivariate_normal distribution essentially draws random samples from a multivariate normal distribution. The multivariate normal distribution is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its **mean** and **covariance** matrix. These parameters are analogous to the mean (average or “centre”) and variance (standard deviation, or “width,” squared) of the one-dimensional normal distribution [3]. 

To fulfil the requirements of the first parameter, we can specify the means of each variable as outputted in the descriptives table above, and then input these values into the first multivariate_normal parameter space. For the second covariance parameter, numpy offers  popular approximations such as spherical or diagonal covariance which can be used. Nevertheless, in this case it is preffered that we simulate an approximate "guess-like" covariance matrix of our own, one that is intended to reflect the true relationships between the student variables:

```python
# Formatting the covariance parameter matrix based on approximation of the true values [7]:
# e.g. The correlation of the Age variable with itself is 1, and 0.5 is a "guessed" covariance value between Age and Sex (first row, second value), .

cov = np.array([[1, 0.5,0.4,0.2,0.3,0.4,0.7,0.4],
                [0.5,1,0.2,0.3,0.3,0.9,0.5,0.2],
                [0.4,0.2,1,0.9,0.7,-0.8,0.5,0.5],
                [0.2,0.3,0.9,1,0.3,-0.4,0.5,0.6], 
                [0.3,0.3,0.7,0.3,1,0.4,0.3,0.3],
                [0.4,0.9,-0.8,-0.4,0.4,1,-0.5,-0.2],
                [0.7,0.5,0.5,0.5,0.3,-0.5,1,0.3],
                [0.4,0.2,0.5,0.6,0.3,-0.2,0.3,1]])
```

```python
cov # Outputting the simulated covariance values
```

Next, for the mean parameter, we take the mean value for each student variable from our descriptive table above and input these into the first random.multivariate_normal parameter space. Following this we add our estimated covariance values into the second parameter space, and lastly the size parameter of 250 students:

```python
import warnings # Importing the Python warnings module to ignore a benign "ipykernel package" warning
warnings.filterwarnings('ignore')

np.random.seed(1234)

# Inputting the mean score and covariance for each student variable:

stu_scores = np.random.multivariate_normal(mean = [41.,0.5,65,12,0.5,12,48,686], cov=cov, size=250)

# Next we generate the updated dataframe using random.multivariate_normal so that we can conduct correlational analyses:

dfnewer = pd.DataFrame(stu_scores, columns = {'Age':Age.round(0), 'Sex':np.random.randint(1, 3, 250), 'HDip_grade':np.random.normal(65, 12,250).round(1), 'Hours_per_week_study':np.random.normal(12, 4, 250).round(2) , 'Course_type':np.random.binomial(1, 0.5, 250),
                      'Weekly_alcohol_units':np.random.normal(12, 5, 250).round(), 'Psychological_resilience':np.random.uniform(0, 100, 250).round(2), 'Total_Moodle_logins':np.random.logistic(700, 60, 250).round(0)})


# Before we generate the correlation table, we should remove the underscores in some of the variable names:

dfnewer = dfnewer.rename(columns={'HDip_grade': 'HDip grade', 'Hours_per_week_study': 'Hours per week study', 
                                  'Course_type': 'Course type', 'Weekly_alcohol_units': 'Weekly alcohol units', 
                                 'Psychological_resilience': 'Psychological resilience', 'Total_Moodle_logins': 'Total Moodle logins' }) # Renaming columns to remove underscores

warnings.resetwarnings() # Reset the previously ignored warning to its default state

dfnewer.corr().round(2) # Generate the correlation table:
```

The correlation table above shows a wide variety of r values (coefficients). The coefficient r measures the strength and direction of a linear relationship between two variables, typically on a scatterplot. As we can see, some values are negative while most are positive. Rumsey (2016) offers a neat method for interpreting the values. For example, a coefficient of "0" means absolutely no linear relationship is present [11]:

* **–1.00**: A perfect downhill (negative) linear relationship
* **–0.70**: A strong downhill (negative) linear relationship
* **–0.50**: A moderate downhill (negative) relationship
* **–0.30**: A weak downhill (negative) linear relationship

*  **0.00**: No relationship at all

* **+0.30**: A weak uphill (positive) linear relationship
* **+0.50**: A moderate uphill (positive) relationship
* **+0.70**: A strong uphill (positive) linear relationship
* **+1.00**: A perfect uphill (positive) linear relationship

#### Exploratory analysis

Aside from looking at relationships between the student variables, we can also conduct some exploratory analyses. 

For example, it would be interesting to see if there are any significant differences between certain groups, e.g. male vs female Moodle logins, or if Data Analytics students consume more alcohol per week than Software Developers. To investigate if any such differences exist, we can conduct a T-test. 

We can use a T-test, if we observe two independent samples from the same or different population, which in this case is true, i.e. males and females or the two course types. The T-test measures whether the average (expected) value differs significantly across the group samples. Like the previous regression tests, if we observe a large p-value, e.g. larger than 0.05 or 0.1, then we cannot reject a null hypothesis of identical average scores. If the p-value is smaller than the threshold, e.g. 1%, or 5%, then we can reject the null hypothesis [14]. Let's investigate the grades among course type first: 

```python
# First we need to import ttest_ind from SciPy as our groups are independent [13]:

from scipy.stats import ttest_ind 

Data_Analytics = dfnewer[dfnewer['Course type']=='Data Analytics']
Software_Dev = dfnewer[dfnewer['Course type']=='Software Dev']

# Output the T statistic and p value:

ttest_ind(Data_Analytics['HDip grade'], Software_Dev['HDip grade'])
```

Very interesting, but not a significant outcome. Looking at the p value output we can see it is not significant but still quite low. Let's see what the average grades actually were by group:

```python
dfnewer.groupby(['Course type'])['HDip grade'].mean().round(1)
```

Almost identical. Now let's look at another group. It would be interesting to see if there exists any difference in the weekly alcohol consumption of males and female students:

```python
Male = dfnewer[dfnewer['Sex']=='Male']
Female = dfnewer[dfnewer['Sex']=='Female']

# Output the T statistic and p value:

ttest_ind(Male['Weekly alcohol units'], Female['Weekly alcohol units'])
```

As we can see from the p value is very high, thus alluding to no significant difference between male and female students and how much alcohol they consume weekly. Let's find out exacly:

```python
dfnewer.groupby(['Sex'])['Weekly alcohol units'].mean().round(1)
```

No difference at all here. It is always good practice to investigate the categorical variables in a data set against the continuous variables. There may be something meaningful or noteworthy in the data that is difficult to spot initially.

#### Basic supervised machine learning on students grades

We have investigated some of the relationships and differences that exist among student variables. It would be interesting to do something a little different with the data. We could use KNN classification to try to predict student "grade range", i.e. first class, second class, third class honours or a fail. It would be interesting to see if grade range can be predicted on the basis of Hours per week study, Psychological resilience and total Moodle logins. 

**Classification** essentially identifies to which of a set of categories a *new* observation belongs, on the basis of a training set of data containing observations (instances) whose category membership is already known. Thus, with our classification algorithm, we should be able to enter values of any new students, i.e. their study hours, resilience and Moodle logins and then, potentially predict whether they are on course to attaining a first class honours or to possibly failing their course.

To do this we will first create a new categorical variable and add it to the dataframe. This new variable will be called "Grade range". We will create the four grade categories using a *for* loop that iterates over the pre-existing HDip grade variable. Following this, we can begin building the KNN classifier [16].

```python
# Create a new grade range list to store the data [15]:
grades = []

# For each row in the column:
for row in dfnew['HDip_grade']:
    
    if row >= 70 < 100:
        # Append a letter grade
        grades.append('1.1-First class honours')
    # else, if more than a value,
    elif row >= 60 <70:
        # Append a letter grade
        grades.append('2.1-Second class honours')
    # else, if more than a value,
    elif row >= 50 <60:
        # Append a letter grade
        grades.append('2.2-Third class honours')
    # otherwise,
    else:
        # Append a failing grade
        grades.append('Failed')
        
        
# Create a column from the list:
dfnew['Grade range'] = grades
grades
```

```python
dfnew # Eyeballing our new column
```

```python
# Import the necessary libraries:

import sklearn.neighbors as nei
import pandas as pd
import sklearn.model_selection as mod

# First we need to train a clasifier on the dataset by setting up the datapoint inputs:
# Inputs and outputs:
inputs = dfnew[['Hours_per_week_study', 'Psychological_resilience', 'Total_Moodle_logins']] # selecting a sub selection of variables from the dataframe
outputs = dfnew['Grade range']

# Classifier:
# Now we create an "object" that represents the KNN algorithm where we can set the value of K
# KNN works by a "Majority vote" method 
knn = nei.KNeighborsClassifier(n_neighbors=2) # We make an object that represents the KNN algo where K is set at 2
                                             
# Fitting:
# Setting up the datapoints that will be used to build a model, to make predictions, i.e. fitting the data to the model (training step)
# knn.fit (method) has now had the inputs and outputs fitted to it.

knn.fit(inputs, outputs)
```

```python
# Predict:
# Now we can test out a random row example, to see if the KNN predicts correctly

dfnew.loc[30]
```

```python
# knn.predict receives sets of inputs to predict an output:

knn.predict([[12.25, 61.05, 608]]) # Checking the example with our KNN predictor
```

The result is correct, now we can input a couple of novel student values to see what grade the new student might attain:

```python
knn.predict([[18, 40, 900]]) # Enter to make prediction 1:Hours per week study, 2:Psychological resilience, 3:Total Moodle logins
```

```python
knn.predict([[2, 70, 500]]) # Enter to make prediction 1:Hours per week study, 2:Psychological resilience, 3:Total Moodle logins
```

Interesting, even though the first student had a lower measure of psychological resilience (40), he/she still attained a first class honours. 

Now we can evaluate the model to determine how accurate the predictions really are: 

```python
# Evaluate the model:
# Comparing inputs (numpy array) to the grades output (data series)

(knn.predict(inputs) == outputs)
```

```python
# Using sum function to show how many comparisons are true:
# By convention, Pandas will convert True to 1s and False to 0s.

(knn.predict(inputs) == outputs).sum()
```

Why is the classifier predicting so many wrong? Possibly because the closest points (neighbours) to the trained values are actually outweighing those value and thus resulting in it being incorrectly classified. Interesting but not very reliable after all. Therefore, to attempt to overcome these miscalculations we can take a random selection and train and test them. Here we will pass in inputs, outputs and test size ratio.

```python
# Train/test split method:
# Takes the dataframe, splits it, e.g. trains on 125 student grades and tests this on the remaining 125 (inputs/outputs test). 
# In scikit X means inputs, y means outputs.
# The inputs and outputs trained (and tested) are in fact allinged when splitted (thus corresponding).
# This uses numpy random for a different outcome everytime.
# The train/test split is always random and thus won't always yield the same results.

inputs_trained, inputs_tested, outputs_trained, outputs_tested = mod.train_test_split(inputs, outputs, test_size=0.50)
```

Now we will "recreate" the classifier and train it on the randomly selected inputs and outputs.

```python
# New classfier (recreated), fit new classifier, to train up model:  

knn = nei.KNeighborsClassifier(n_neighbors=2)
knn.fit(inputs_trained, outputs_trained)
```

```python
(knn.predict(inputs_tested) == outputs_tested).sum()
```

Only 49 of the 125 inputs are true. This outputted value would typically lead us to infer that the model is not very accurate at predicting student grade ranges. Why might this be? 

There are many reasons why a model's accuracy might be weaker than expected. For example, it could be down to the random effect of the train/test split on the classifier, or train/test may have selected a not so great training set out of the 125 examples. The KNN doesn't always get things correct, even when it could typically come down to how random the data itself is. Indeed, sometimes the machine learning algorithm is not such a good fit for the data or vice versa, and thus warrants more investigation. Alternatively, there may in fact be an effect in the data, but the KNN doesn't pick it up. Carefully cleansing the data prior to analysis is good practice before applying any statistical tests or machine learning techniques. Assessing accuracy is also important, but there can be other factors that can contribute to a model being weaker than desired. For example, the combined predicting variables may not be that reliable at predicting grade range *together*. As we have seen from the previous regression plots, Hours per week studied was actually an excellent predictor of grade, but when the extra predictor variables (resilience and Moodle logins) were added on top of it, the effect was notably weekened. 

So, that particular model was not ideal to use with this particular data set. We have other options. For example, we could try using a simple decision tree classifier. Here we could again try to predict Grade range based on slighly different variables. A decision tree is a decision tool. Its similar to a tree-like model in computer science. (root at the top, leaves downwards). Again, this is a supervised learning technique, we feed the classifier with inputs from the existing student data to "learn" (or train) from, then we can can make new predictions on what a students Grade range will be based on those learned inputs [25]:

```python
# Import tree from sklearn and create the model [25]:

from sklearn import tree
clf = tree.DecisionTreeClassifier()   
```

```python
# Create the training data for the classifier/decision tree:

X = dfnew[['Hours_per_week_study', 'Weekly_alcohol_units', 'Psychological_resilience']] # Inputs for the classifier to learn from

Y = dfnew['Grade range'] # Outputted "decision" based on learned inputs
```

```python
# Putting it all together [25]:

clf = clf.fit(X, Y) # Fitting the model 

# Now we can test out a random row example, to see if the tree predicts correctly

dfnew.loc[70]
```

```python
# clf.predict receives sets of inputs to predict an output:

clf.predict([[9.97, 29, 49.33]]) # Checking the example with our tree predictor
```

So the classifier predicted correctly that the grade range was a fail. Not a nice outcome for any student. This student was female and of the data analytics group, she commited almost 10 hours a week to studying, consumed an average of 29 units of alcohol and was quite low in resilience (49). We can only tentatively speculate about what happened, but academia seemed to get the better of her. Perhaps this student should have tried cut down a little on the alcohol intake and focus a bit more. Let's input some data belonging to a new student to see what the prediction is:

```python
# Making a prediction based on new student data:

new_prediction = clf.predict([[10, 2, 80]])  # Inputting novel student data values for prediction (Hours per week study, Weekly alcohol units, Psychological resilience)                                      
new_prediction
```

As we can see above, based on our tree classifier, a new student who studies for **10** hours, consumes **2** units of alcohol per week and has a resilience score of **80** will attain a first class honours. Great for him or her.
