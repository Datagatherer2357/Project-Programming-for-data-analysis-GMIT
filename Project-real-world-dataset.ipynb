### The phenomenon

Psychological resilience is an essential and valuable trait for any third level student to possess. Whilst third level education is meant to be a very exciting and enjoyable time in a persons life, students endure some of the most gruelling and stressful academic challenges, especially from an emotional and psychological point of view. Indeed, the manner in which a student copes with the pressures of working toward attaining a first class undergraduate or postgraduate qualification will essentially have notable effects on their own mental wellbeing and final year grade. Every student employs their own coping styles, whether it's meditation, watching Netflix, listening to music, talking to friends or even alcohol consumption. The way in which each student copes, whether they adopt adaptive or maladaptive strategies will ultimately have reciprocal effects on their academic performance, and essentially their emotional and psychological wellbeing. But what exactly is psychological resilience and how does it relate to student performance?

#### Pychological resilience

Psychological resilience can be defined as the ability to successfully cope with a crisis and to return to pre-crisis status quickly [1]. Resilience is apparent when the person uses mental processes and behaviours in promoting personal assets and protecting oneself from the potential negative effects of stressors. Essesntially, psychological resilience exists in people who develop psyhological and behavioural capabilities that allow them to reamin calm and objective during crises and to move forward from these events without long-term negative consequences. Resilience is a positive adaption after a stressful or adverse situation. When a person is overwhelmed by daily stress, it disrupts their internal and external sense of balance which can present both challenges and opportunities. It is not just about overcoming deeply sressful situations but also about emerging from them with competent functioning. Resilience is best understood as a process more than a trait. It is a cycle, a reactional process to stressful situations followed by an adaptive or maladaptive behaviour. Indeed, stressors can cause students to erupt with anger, implode with negative emotions or go numb, or simply become upset about the disruptive change. The first two reactions tend to lead people to play the victim by blaming others and rejecting any coping methods. These individuals instinctively *react* rather than *respond*. The third reaction type promotes wellbeing and is employed by resilient people who become upset about the disruptive state, yet change their current position to cope with the issue. These individuals tend to step back, make realistic plans, have confidenece in their own strengths, have good problem solving skills and have a higher degree of emotional regulation than those who choose to submit to natural reactions of fear, anger and helplessness [2].

This study aims explore some of the most important variables including psychological resilience, which affect and relate to student performance in an Irish Institute of Technology. These students have recently completed their Higher Diplomas cand are from two different course cohorts: Those who studied Data Analtics and those who studied Software Development.

Let's first take a look at the Age variable in the dataset. The Age variable is a continuous numerical variable. For both cohorts of students their ages range from 17 to 65. Age distributions can vary depending on the cohorts in the study. For example most population pyramids will display a trapezoidal or pyramid shaped distibution. For our 250 students ranging from 17 to 65 years, I felt a blend of uniform and triangular distribution would best reperesent the samples. Both  distribution parameters allow you to set both the lower and upper ranges of age, but the triangular distribution allows a  mode, i.e. the value which appears most often (the peak) to be set as well [3].

```python
# Import the necessary modules:

import pandas as pd
import numpy as np

# Set the seed so the variable values are reproducible:

np.random.seed(1234) 


# Setting the age range and sample size of students by concatenating the uniform and triangular distributions.
# This method will generate a convincing student age distribution [3]. 

Age1 = np.random.uniform(17,65,125)  # (low, high, size)
Age2 = np.random.triangular(17,37,65,125) # (left, mode, right, size)
Age = np.concatenate((Age1,Age2)) # Concatenating the distributions

Age.round(0)
```

```python
len(Age) # Checking the count of our newly concatenated distribution
```

Now let's plot a histogram which shows both merged uniform and triangular distributions of Age:

```python
import matplotlib.pyplot as plt 
%matplotlib inline

# Configuration of histogram [6]:

bins = np.linspace(0, 70, 30) # linspace returns evenly spaced numbers over a specified interval
plt.hist([Age1, Age2], bins, label=['Age1', 'Age2']) # Adding distribution samples and labelling them
plt.legend(loc='upper left') # Legend location
plt.ylabel('Frequency') # labelling y axis
plt.xlabel('Age') # labelling x axis 
plt.title('Student age') # Adding title

plt.show()
```

As we can see from the double histogram, the orange (Age2) triangular distribution has a very high peak giving it its shape. The uniform distribution is still quite similar to the triangular apart from the fact it lacks the kurtosis feature. Taken as one complete distribution, we now have an adequate representation of student age.

Next we should explore some relationships between the student variables. For this we need to look at the covariance values of the student variables. 

Covariance and correlation will describe how two variables are related in our dataset. Essentially, the student variables are infered as positively related if they move in the same direction, or negatively related if they move in opposite directions. Covariance indicates how two variables are related, i.e. positively or negatively, and it measures variables that have different *units* of measurement. However, it is not possible to measure the *degree* to which the variables move together because covariance does not use one standard unit of measurement. Therefore, to measure the degree to which variables move together, we use correlation. Correlation standardizes the measure of *interdependence* between any two variables and, consequently, tells us how closely the two variables move. The correlation measurement (coefficient), always takes on a value between 1 and – 1 [9]. 

In order to calculate and evaulate the correlations among the student variables, we first need to generate covariance values that offer a close approximate simulation of the true relationships of the variables. To do this we stay with the numpy.random packages and employ the numpy.random.multivariate_normal distribution function. The multivariate normal distribution is often used to describe, at least approximately, any set of (possibly) correlated real-valued *random variables* each of which clusters around a mean value [10]. Therefore, this distributional function is an ideal candidate for the job of simulating covariance values and then determining the correlations between the student variables.

The random.multivariate_normal distribution draws random samples from a multivariate normal distribution. The multivariate normal distribution is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and covariance matrix. These parameters are analogous to the mean (average or “center”) and variance (standard deviation, or “width,” squared) of the one-dimensional normal distribution [3]. 

We can specify the means of each variable as outputted in the descriptives table into the first multivariate_normal parameter. Regarding the required covariance parameter, numpy offers  popular approximations such as spherical or diagonal covariance which can be used. Nevertheless, in this case it is preferred we create an approximate "guess-like" covariance matrix of our own, intending to reflect hypothesized relationships between the variables. Let's take a look:

```python
# Formatting the covariance parameter matrix based on approximation [7]:
# e.g. The correlation of the Age variable with itself is 1, and 0.5 is a guessed covariance value between Age and Sex (first row, second value), .

cov = np.array([[1, 0.5,0.4,0.2,0.3,0.4,0.7,0.4],
                [0.5,1,0.2,0.3,0.3,0.4,0.5,0.2],
                [0.4,0.2,1,0.9,0.7,-0.6,0.5,0.5],
                [0.2,0.3,0.9,1,0.3,-0.4,0.5,0.6], 
                [0.3,0.3,0.7,0.3,1,0.4,0.3,0.3],
                [0.4,0.4,-0.6,-0.4,0.4,1,-0.5,-0.2],
                [0.7,0.5,0.5,0.5,0.3,-0.5,1,0.3],
                [0.4,0.2,0.5,0.6,0.3,-0.2,0.3,1]])
```

```python
cov # Outputting the approximate covariance values
```

Next, for the mean parameter we input the mean score for each student variable taken from our descriptive table above into the random.multivariate_normal parameters. Following this we add our estimated covariance values into the second parameter, and lastly the size parameter of 250 students:

```python
import warnings # Importing the Python warnings module to ignore a benign "ipykernel package" warning
warnings.filterwarnings('ignore')

np.random.seed(1234)

# Inputting the mean score and covariance for each student variable:

stu_scores = np.random.multivariate_normal(mean = [41.,0.5,65,12,0.5,12,48,686], cov=cov, size=250)

# Next we generate the updated dataframe using random.multivariate_normal so that we can conduct correlational analyses:

dfnew = pd.DataFrame(stu_scores, columns = {'Age':Age.round(0), 'Sex':np.random.randint(1, 3, 250), 'HDip_grade':np.random.normal(65, 12,250).round(1), 'Hours_per_week_study':np.random.normal(12, 4, 250).round(2) , 'Course_type':np.random.binomial(1, 0.5, 250),
                      'Weekly_alcohol_units':np.random.normal(12, 5, 250).round(), 'Psychological_resilience':np.random.uniform(0, 100, 250).round(2), 'Total_Moodle_logins':np.random.logistic(700, 60, 250).round(0)})


# Before we generate the correlation table, we should remove the underscores in some of the variable names:

dfnew = dfnew.rename(columns={'HDip_grade': 'HDip grade', 'Hours_per_week_study': 'Hours per week study', 
                                  'Course_type': 'Course type', 'Weekly_alcohol_units': 'Weekly alcohol units', 
                                 'Psychological_resilience': 'Psychological resilience', 'Total_Moodle_logins': 'Total Moodle logins' }) # Renaming columns to remove underscores

warnings.resetwarnings() # Reset the previously ignored warning to its default state

dfnew.corr().round(2) # Generate the correlation table:
```

Looking at the correlation table above, we can see a wide variety of r values (coefficients). The coefficient r measures the strength and direction of a linear relationship between two variables, typically on a scatterplot. As we can see, some values are negative while most are positive. Rumsey (2016) offers a neat method for interpreting the values. For example, a coefficient of "0" means absolutely no linear relationship is present [11]:

* **–1.00**: A perfect downhill (negative) linear relationship
* **–0.70**: A strong downhill (negative) linear relationship
* **–0.50**: A moderate downhill (negative) relationship
* **–0.30**: A weak downhill (negative) linear relationship

*  **0.00**: No relationship at all

* **+0.30**: A weak uphill (positive) linear relationship
* **+0.50**: A moderate uphill (positive) relationship
* **+0.70**: A strong uphill (positive) linear relationship
* **+1.00**: A perfect uphill (positive) linear relationship
