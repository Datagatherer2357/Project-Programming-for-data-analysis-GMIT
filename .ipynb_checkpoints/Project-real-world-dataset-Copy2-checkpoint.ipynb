```python
# Age with datafram

import pandas as pd
import numpy as np

#np.random.seed(42) # So the variable values are reproducible

df = pd.DataFrame({'Student ID': np.random.randint(12345, 13432, 250), 
                   'Age':np.random.randint(17,65, 250), 'Sex':np.random.randint(1, 3, 250), 'HDip grade':np.random.normal(60, 14,250).round(1)})
df
```

```python
# Replacing sex values to strings!!
# https://stackoverflow.com/questions/31888871/pandas-replacing-column-values

df['Sex'].replace({1:'Male', 2:'Female'}, inplace=True)
df
```

```python
# Adding a new variables column: 
# https://erikrood.com/Python_References/add_new_col_df_default_value_final.html

df['Hours per week study'] = np.random.normal(10, 5, 250)

df = df.round(2) # Rounding the decimal places 
#df.drop('My old column', axis=1, inplace=True) # Removing a column
#df = df.rename(columns={'Hdip_grade': 'HDip grade'}) # Renaming a column
df
```

```python
df['Course type'] = np.random.randint(0, 2, 250) # Adding course type variable
df['Course type'].replace({0:'Data Analytics', 1:'Software Dev'}, inplace=True) # Coverting numerics to strings
#df.drop('HDip grade', axis=1, inplace=True) # Removing a column
df
```

```python
# Adding the alcohol intake variable and rounding it to 0 decimals

df['Weekly alcohol units'] = np.random.normal(10, 5, 250).round()
df
```

```python
df['Psychological reslience'] = np.random.logistic(55, 7, 250).round(0) # Created with logistic distribution. It resembles the normal distribution in shape but has heavier tails (higher kurtosis).
df
```

```python
df['HDip grade'].corr(df['Weekly alcohol units'])
```

```python
df.corr() # Shows how wrongly distributed the random data is...needs to be fixed
```

```python
# factorisation alternatives to correlate categorical variables
# https://stackoverflow.com/questions/47894387/how-to-correlate-categorical-column-in-pandas

df['Course type'] = df.groupby('Course type').ngroup()
df['Sex'] = df.groupby('Sex').ngroup()
df
```

```python
# So, I've created the basic data, now lets convert them and make them correlate...Awesome
# https://stackoverflow.com/questions/45952895/generating-correlated-numbers-in-numpy-pandas
# Explain all of this code and reference it!

# 1 construct covariance data:
# 0.5 is the correlation between A and B, and the correlation of a variable with itself is 1.0.
# The arrays belowe MUST mirror each other !! AMEND
cov = np.array([[1, 0.5,.4, .2, .3, .4, .5],[.5,1.,.1,.1,3,4,5],[0.2,.1,1,.3,3,4,5],[0.1,.1,.3,1,3,4,5], [.5,1.,.1,.5,1,4,5],[0.2,.1,1,.3,3,1,5],[0.7,.1,.3,1,3,4,1]])

# 2  generate data of our 250 points:
scores = np.random.multivariate_normal(mean = [29.,1.5,60,8,0.5,5,40], cov=cov, size = 250)

# 3  let's put my data into a Pandas DataFrame:
cf = pd.DataFrame(scores, columns = ["Age","Sex","HDip grade", "Hours per week study", "Course type", "Weekly alcohol units", "Psychological resilience"])
cf.head(250)
```
